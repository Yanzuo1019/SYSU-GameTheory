```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
avg_rewards1 = []
avg_rewards2 = []
with open("DQN/DQN-result.txt", "r") as f1:
    with open("PER-Dueling-DQN/PER-Dueling-DQN-result.txt", "r") as f2:
        for line1, line2 in zip(f1.readlines(), f2.readlines()):
                avg_rewards1.append(float(line1.strip()))
                avg_rewards2.append(float(line2.strip()))
```

```python
avg_rewards3 = []
with open("Dueling-DQN-reward.txt", "r") as f3:
    for line in f3.readlines():
        avg_rewards3.append(float(line.strip().split()[2]))
```

```python
avg_rewards1 = np.array(avg_rewards1)
avg_rewards2 = np.array(avg_rewards2)
accu_rewards1 = []
accu_rewards2 = []
for i in range(50, 500):
    accu_rewards1.append(np.mean(avg_rewards1[i-50:i]))
    accu_rewards2.append(np.mean(avg_rewards2[i-50:i]))
```

```python
avg_rewards3 = np.array(avg_rewards3)
accu_rewards3 = []
for i in range(5, 50):
    accu_rewards3.append(np.mean(avg_rewards3[i-5:i]))
```

```python
plt.plot(range(50, 500), accu_rewards1)
plt.plot(range(50, 500), accu_rewards2)
plt.plot(range(50, 500, 10), accu_rewards3)
plt.ylabel("last 50 models average rewards")
plt.legend(["DQN", "PER-Dueling-DQN", "Dueling-DQN"])
plt.savefig("result.jpg")
plt.show()
```
